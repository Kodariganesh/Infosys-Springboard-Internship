```python
import numpy as np
def assign_category(row):
    if "billing" in row['subject'].lower() or "billing" in row['body'].lower():
        return "Billing Inquiries"
    elif "critical" in row['subject'].lower() or "urgent" in row['tag_1'].lower():
        return "Urgent Issues"
    elif "sales" in row['subject'].lower() or "sales" in row['tag_4'].lower():
        return "Sales Inquiries"
    elif "support" in row['queue'].lower():
        return "Technical Support"
    else:
        return "General Inquiry"
df['category'] = df.apply(assign_category, axis=1)
```

```python
df.head(2)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(df['answer'])
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out())
tfidf_df.head()
```

```python
(tfidf_df == 0.0).sum(axis=1).sum() / (tfidf_df.shape[0] * tfidf_df.shape[1]) * 100
```

```python
tfidf_df.shape
```

```python
top_keywords = tfidf_df.sum().sort_values(ascending=False).head(10)
top_keywords
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(top_keywords)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
print("Shape of TF-IDF Matrix:", tfidf_matrix.shape)
```

```python
sparsity = 1.0 - (tfidf_matrix.nnz / float(tfidf_matrix.shape[0] * tfidf_matrix.shape[1]))
print("Sparsity:", sparsity)
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

pca = PCA()
pca.fit(tfidf_matrix.toarray())
explained_variance = pca.explained_variance_ratio_

plt.figure(figsize=(8, 5))
plt.plot(range(1, len(explained_variance) + 1), explained_variance.cumsum(), marker='o', linestyle='--')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('PCA - Explained Variance')
plt.grid(True)
plt.show()
```

```python

from sklearn.decomposition import PCA
pca = PCA(n_components=3)
pca.fit(tfidf_df)
principal_components = pca.transform(tfidf_df)
pca_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2', 'PC3'])

pca_df.head()
```

```python
(pca.explained_variance_ratio_ * 100).sum()
```

```python

import matplotlib.pyplot as plt
n_components = min(tfidf_df.shape[0], tfidf_df.shape[1])
n_components = min(n_components, 10)

pca = PCA(n_components=n_components)
pca.fit(tfidf_df)

principal_components = pca.transform(tfidf_df)

plt.figure(figsize=(10, 6))
plt.plot(range(1, n_components + 1), pca.explained_variance_ratio_[:n_components] * 100, marker='o', linestyle='-')
plt.title('Top {} Principal Components Variance Ratio'.format(n_components))
plt.xlabel('Principal Component')
plt.ylabel('Variance Ratio (%)')
plt.grid(True)
plt.show()
```

```python

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
inertias = []
for k in range(1, 15):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(tfidf_df)
    inertias.append(kmeans.inertia_)
plt.figure(figsize=(8, 6))
plt.plot(range(1, 15), inertias, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()
```

```python

from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
inertias = []
for k in range(1, 15):
    agglomerative = AgglomerativeClustering(n_clusters=k)
    agglomerative.fit(tfidf_df)

    labels = agglomerative.labels_
    inertias.append(len(labels))
plt.figure(figsize=(8, 6))
plt.plot(range(1, 15), inertias, marker='o')
plt.title('Elbow Method (Agglomerative Clustering) for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Number of Samples per cluster')
plt.grid(True)
plt.show()
```

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
pca.fit(tfidf_df)
principal_components = pca.transform(tfidf_df)
pca_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2', 'PC3'])
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

scatter = ax.scatter(pca_df['PC1'], pca_df['PC2'], pca_df['PC3'], c=pca_df['PC3'], cmap='viridis')
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
plt.title('3D Scatter Plot of Principal Components')
fig.colorbar(scatter)
plt.show()
```
