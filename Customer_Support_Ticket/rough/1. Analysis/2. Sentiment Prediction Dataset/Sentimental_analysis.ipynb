```python
import requests
import json

def get_sentiment_gork(title, description):
    url = "https://api.x.ai/v1/chat/completions"
    headers = {
        "Content-Type": "application/json",
        "Authorization": "Bearer xai-4AXGgHhJGcZnClfWiD5JtilLQHnkLfQmaOKT9shwCCm5nRh0SsnvBnRy43ADNudfEek7aoXe09PScPql"

    messages = [
        {
            "role": "system",
            "content": """You are a Customer Support Agent. You have to decide sentiment of the given input, using:
1. Title
2. Chat History

Analyze and provide sentiment based on the inputs.
"""
        },
        {
            "role": "user",
            "content": f"""Title: "{title}"\n\n\nChat History: "{description}" """
        }
    ]
    data = {
        "model": "grok-beta",
        "messages": messages,
        "temperature": 0.5
    }

    try:

        response = requests.post(url, json=data, headers=headers, verify=False)
        print("Full API Response:", json.dumps(response.json(), indent=2))

        if response.status_code == 200:
            chat_completion = response.json()
            if 'choices' in chat_completion and chat_completion['choices']:
                choice = chat_completion['choices'][0]
                sentiment_content = choice.get('message', {}).get('content', '')
                sentiment = 'neutral'
                if 'positive' in sentiment_content.lower():
                    sentiment = 'positive'
                elif 'negative' in sentiment_content.lower():
                    sentiment = 'negative'
                elif 'neutral' in sentiment_content.lower():
                    sentiment = 'neutral'
                sentiment_result = {
                    "sentiment": sentiment,
                    "thought": sentiment_content
                }

                return sentiment_result
            else:
                print("No choices found in response.")
                return {"error": "No sentiment analysis found."}
        else:
            print(f"Error: {response.status_code}, {response.text}")
            return {"error": f"Request failed with status {response.status_code}"}
    except requests.exceptions.RequestException as e:
        print(f"An error occurred: {e}")
        return {"error": str(e)}
```
