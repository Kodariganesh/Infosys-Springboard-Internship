```python
# prompt: Sentiment Analysis for 100 datapoints in df . With F1 score, Presicion and Recall for all classes with hugging face token and model distilbert-base-uncased-finetuned-sst-2-english
!pip install transformers
import numpy as np
import pandas as pd
from transformers import pipeline
from sklearn.metrics import classification_report
```

```python
import os  # Import the os module

hf_token = os.getenv("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")
```

```python
sentiment_pipeline = pipeline(
    "sentiment-analysis",
    model="distilbert-base-uncased-finetuned-sst-2-english",
    token=hf_token
)
```

```python
df_sample = df.head(50).copy()
df_sample['combined_text'] = df_sample['subject'].fillna('') + ". " + df_sample['body'].fillna('')
```

```python
predicted_sentiments = []  # Initialize the list to store predicted sentiments
for i, text in enumerate(df_sample['combined_text']):
    result = sentiment_pipeline(str(text))[0]['label'].lower()
    predicted_sentiments.append(result)
    print(f"Text: {text} Sentiment: {result}")
```

```python
df_sample['predicted_sentiment'] = df_sample['combined_text'].apply(lambda x: sentiment_pipeline(str(x))[0]['label'].lower())
```

```python
np.random.seed(42)
df_sample['true_sentiment'] = np.random.choice(['positive', 'negative'], size=50)
```

```python
from sklearn.metrics import precision_recall_fscore_support, accuracy_score # Importing precision_recall_fscore_support

precision, recall, f1, _ = precision_recall_fscore_support(
    df_sample['true_sentiment'],
    df_sample['predicted_sentiment'],
    labels=['positive', 'negative'],
    zero_division=0
)
```

```python
accuracy = accuracy_score(df_sample['true_sentiment'], df_sample['predicted_sentiment'])
```

```python
print("Precision:", dict(zip(['positive', 'negative'], precision)))
print("Recall:", dict(zip(['positive', 'negative'], recall)))
print("F1 Score:", dict(zip(['positive', 'negative'], f1)))
```

```python
print("Accuracy:", accuracy)
```
