```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer

nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt_tab')

def preprocess_text(text):
    # Tokenize the text
    tokens = nltk.word_tokenize(text.lower())

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words]

    # Lemmatize the tokens
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token) for token in tokens]

    return " ".join(tokens)
```

```python
preprocessed_tickets = df['tickets'].apply(preprocess_text)
```

```python
import pandas as pd
from transformers import AutoTokenizer, AutoModelForTokenClassification, pipeline

# Load pre-trained model and tokenizer
tokenizer = AutoTokenizer.from_pretrained("dslim/bert-base-NER")
model = AutoModelForTokenClassification.from_pretrained("dslim/bert-base-NER")
ner_pipeline = pipeline("ner", model=model, tokenizer=tokenizer)

# Function to extract Named Entities (NER)
def get_ner(text):
    if not text or not isinstance(text, str):
        return []
    return ner_pipeline(text)

# Parse BERT NER results to extract complete tokens
def parse_bert_ner_results(results):
    nouns = []
    current_noun = ""
    last_index = -1

    for entity in results:
        word = entity["word"].replace("##", "")
        if last_index == -1 or entity["index"] != last_index + 1:
            if current_noun:
                nouns.append(current_noun)
            current_noun = word
        else:
            current_noun += f" {word}"
        last_index = entity["index"]

    if current_noun:
        nouns.append(current_noun)
    return nouns

# Function to process a single ticket (JSON input)
def process_ticket(ticket):
    # Convert JSON ticket to DataFrame
    df = pd.DataFrame([ticket])

    # Fill missing text fields
    df["subject"] = df["subject"].fillna("")
    df["body"] = df["body"].fillna("")

    # Apply NER to subject and body
    df["ner_tags_subject"] = df["subject"].apply(get_ner)
    df["ner_tags_body"] = df["body"].apply(get_ner)

    # Extract product names
    df["product_names_subject"] = df["ner_tags_subject"].map(parse_bert_ner_results)
    df["product_names_body"] = df["ner_tags_body"].map(parse_bert_ner_results)

    # Summary
    categories = df["type"].value_counts(normalize=True) * 100
    product_names = pd.concat([df["product_names_subject"].explode(), df["product_names_body"].explode()]).value_counts(normalize=True) * 100
    issues = pd.concat([df["tag_1"], df["tag_2"]]).value_counts(normalize=True) * 100

    # Return results as dictionary
    return {
        "categories": categories.to_dict(),
        "product_names": product_names.to_dict(),
        "issues": issues.to_dict(),
    }

# Example ticket input (JSON format)
example_ticket = {
    "id": 1234,
    "subject": "Issue with AWS instance setup",
    "body": "The EC2 instance on AWS is not launching properly.",
    "type": "Incident",
    "tag_1": "Technical Support",
    "tag_2": "AWS",
    "priority": "High",
}

# Process the ticket
result = process_ticket(example_ticket)
result
```

```python
import pandas as pd
from transformers import AutoTokenizer, AutoModelForTokenClassification, pipeline

# Load pre-trained NER model and tokenizer
tokenizer = AutoTokenizer.from_pretrained("dslim/bert-base-NER")
model = AutoModelForTokenClassification.from_pretrained("dslim/bert-base-NER")
ner_pipeline = pipeline("ner", model=model, tokenizer=tokenizer)

# NER extraction and parsing functions
def get_ner(text):
    """Extract NER tags from text using the BERT pipeline."""
    if not text or not isinstance(text, str):
        return []
    return ner_pipeline(text)

def parse_bert_ner_results(results):
    """Combine NER tokens into full words."""
    nouns = []
    current_noun = ""
    last_index = -1

    for entity in results:
        word = entity["word"].replace("##", "")
        if last_index == -1 or entity["index"] != last_index + 1:
            if current_noun:
                nouns.append(current_noun)
            current_noun = word
        else:
            current_noun += f" {word}"
        last_index = entity["index"]

    if current_noun:
        nouns.append(current_noun)
    return nouns

# Response generation logic
def generate_response(product_name=None, issue=None):
    """Generate a templated response based on product name and issue."""
    if product_name and issue:
        return f"Thank you for reaching out regarding {product_name}. We acknowledge the reported issue '{issue}' and are actively working on it. Our team will provide an update as soon as possible. If you have further details to share, please let us know."
    elif product_name:
        return f"We understand you are facing an issue with {product_name}. Our support team is actively reviewing the problem and will get back to you with a resolution shortly."
    elif issue:
        return f"The reported issue '{issue}' is being prioritized. Please rest assured that we are taking all necessary steps to resolve it at the earliest."
    else:
        return "Thank you for your message. Please provide more details about the issue so we can assist you further."

# Function to process a single ticket (JSON input)
def process_ticket(ticket):
    """Process a ticket JSON object and generate a response."""
    # Convert ticket to DataFrame
    df = pd.DataFrame([ticket])

    # Fill missing text fields
    df["subject"] = df["subject"].fillna("")
    df["body"] = df["body"].fillna("")

    # Extract NER from subject and body
    df["ner_tags_subject"] = df["subject"].apply(get_ner)
    df["ner_tags_body"] = df["body"].apply(get_ner)

    # Parse NER results to extract product names
    df["product_names_subject"] = df["ner_tags_subject"].map(parse_bert_ner_results)
    df["product_names_body"] = df["ner_tags_body"].map(parse_bert_ner_results)

    # Extract relevant information
    product_names = pd.concat([df["product_names_subject"].explode(), df["product_names_body"].explode()]).dropna().unique()
    issues = pd.Series([ticket.get("tag_1"), ticket.get("tag_2")]).dropna().unique()

    # Generate response
    product_name = product_names[0] if len(product_names) > 0 else None
    issue = issues[0] if len(issues) > 0 else None

    response = generate_response(product_name=product_name, issue=issue)
    return {
        "ticket_id": ticket.get("id"),
        "product_name": product_name,
        "issue": issue,
        "response": response,
    }

# Example ticket JSON
example_ticket = {
    "id": 1234,
    "subject": "Issue with AWS instance setup",
    "body": "The EC2 instance on AWS is not launching properly.",
    "type": "Incident",
    "tag_1": "Technical Support",
    "tag_2": "AWS",
    "priority": "High",
}

# Process the ticket and generate a response
result = process_ticket(example_ticket)
result
```
